```python
import requests
import pandas as pd
r = requests.get("https://api-fe.skillindiadigital.gov.in/api/jobs/popular")
```

```python
r.status_code
```

```python
a = r.json()
```

```python
a
```

```python
result=[]
for data in a["Data"]["Results"]:
    row={}
    row['JobTitle']=data["JobTitle"]
    row["IndustryName"]=data["IndustryName"]
    row["ContactPersonName"]=data["ContactPersonName"]
    row["ContactPersonMobile"]=data["ContactPersonMobile"]
    row["ContactPersonEmail"]=data["ContactPersonEmail"]
    result.append(row)
df = pd.DataFrame(result)
df.to_csv('skill.csv', index=False)
```
